````python
import numpy as np
import cv2
import matplotlib.pyplot as plt
````

````python
img_path = '/kaggle/input/test-img/Testing_img.jpeg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (256, 256))

plt.title("Original image")
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
````

### **Q1. What is Smoothing. Why we use?**
Image processing, smoothing (also known as blurring) is used to reduce noise and detail in an image. It's a fundamental operation that helps improve the performance of more complex algorithms like edge detection, object recognition, and feature extraction.

1. Noise Reduction
2. Preprocessing image for the Edge detection
3. Image Enhancement

| Function                | Preserves Edges                     | Good for Noise | Fast?     |
|------------------------|-------------------------------------|----------------|-----------|
| `cv2.blur()`            |  No                                |  Limited      |  Fast    |
| `cv2.GaussianBlur()`    |  Partially                         |  Yes          |  Fast    |
| `cv2.medianBlur()`      |  Yes (excellent for salt & pepper) |  Yes          |  Slower |
| `cv2.bilateralFilter()` |  Yes (best)                        |  Yes          |  Slow    |


### **1. Averaging -> ```cv2.blur()```**

````python
blurred = cv2.blur(img, (3, 3))

plt.title("Avg Blurring image")
plt.imshow(blurred, cmap='gray')
plt.axis('off')
plt.show()
````

### **2. Gaussian Blur -> ```cv2.GaussianBlur```**

````python
# Gaussian blur with 5x5 kernel, sigmaX=0 lets OpenCV compute it
gaussian_blur = cv2.GaussianBlur(img, (3, 3), 0)

plt.title("Gaussian Blur")
plt.imshow(gaussian_blur, cmap='gray')
plt.axis('off')
plt.show()
````

### **3. Median Blur -> ```cv2.medianBlur()```**

````python
# Median blur (kernel size must be odd and >1)
median_blur = cv2.medianBlur(img, 5)

plt.title("Median Blur")
plt.imshow(median_blur, cmap='gray')
plt.axis('off')
plt.show()
````

### **4. Bilateral Filter -> ```cv2.bilateralFilter()```**

````python
# Bilateral filtering: smooths but keeps edges
bilateral = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)

plt.title("Bilateral Filter")
plt.imshow(bilateral, cmap='gray')
plt.axis('off')
plt.show()
````
